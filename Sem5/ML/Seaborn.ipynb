-- Seaborn Library

Seaborn library aims to make a more attractive visualisation of the central part of understandin and exploring data.
It is built on the core of matplotlib Library and provides dataset oriented APIs

-- Categories of plots in Seaborn
Plots are generally used to make visulaisation of relationships between the given variables. These variables can be either be a category like division group or class or they can be completely numerical varables.

- Distribution Plot: this type of plot is used for examining both types od distributions, i.e. univariate and bivariate distribution.
- Relational Plot: This type of plot is used to understand the relation between the two given variables.
- Regression Plot: Regression plot in the seaborn library are primarily intended to add an additional visual guide that will help to ephasize dataset    
                   patterns during the analysis of exploratory data
- Categorical plots: The categorical plots are used to deal with categories of variables and how we can visualize them 
- Multi-plot grids: The Multi-plot grids are also a type of plot that is a useful approach to draw multiple instances for the same plot with different
                    subsets of a single dataset.
- Matrix Plots: The amtrix plots are a type of arrays of the scatterplots.


```python
# -- Load Datasets to construct seaborn plots
# To get the dataset names available in the seaborn library, we use get_dataset_names() on the seaborn object. To import Seaborn library we use 'impor seaborn as sns'
import seaborn as sns
print(sns.get_dataset_names())
```

```python
#To import car_crashdataset() we use the funtion load_dataset() which returns a dataframe and head() is used to print the first five rows by default

df=sns.load_dataset('car_crashes')
print(df.head())
```

```python
#-- Styling and Themes in Seaborn --
# We can set the style on Matplotlib plot using Seaborn's set()

df = sns.load_dataset('car_crashes')
plt.scatter(df.speeding,df.alcohol)
plt.show()
```

```python
# We can also set the style for the plot using the set() as sns.set()
df = sns.load_dataset('car_crashes')
plt.scatter(df.speeding,df.alcohol)
sns.set()
plt.show()
```

```python
#Seaborn supports various themses that can make styling the plots easy and saves lot of time using the set_style() we can set any of one the themes available in seaborn library. The themes can be:
# darkgrid
# whitegrid
# dark, white and ticks.
# The default theme for the plot is darkgrid

# Syntax: sns.set_style("theme_name")

df = sns.load_dataset('car_crashes')
plt.scatter(df.speeding,df.alcohol)
sns.set_style("dark")
plt.show()
```

```python
#Seaborn allows to control individual elements of the graph and control the scale of these elements  by using set_context() which takes four preset templates: 
# - Paper
# - Notebook (default)
# - Talk
# - Poster

# Syntax: sns.set_context("theme_name")
df=sns.load_dataset('car_crashes')
plt.scatter(df.speeding,df.alcohol)
sns.set_style("dark")
sns.set_context("talk")
plt.show()
```

-- Seaborn's Plotting Functions

The Seaborn library provides us with relplot() that shows the relatioship between two variables, it provides accesss to several different access level functions. The kind parameter selects the axes-level function:

- To get scatterplot use kind="scatter" and to get a lineplot we use kind="line", the default value for kind parameter is "scatter"

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.relplot(data=tips, x="total_bill", y="tip")
sns.set_style("darkgrid")
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",hue="day")
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",hue="sex",col="day",col_wrap=4)
```

```python
sns.relplot(data=tips,x="size",y="tip",kind="line")
```

```python
sns.set_style("dark")
iris=sns.load_dataset("iris")
sns.kdeplot(iris.loc[(iris['species']=='setosa'),'sepal_length'],color='b',shade=True,label='Setosa')
plt.show()
```

```python
sns.set_style("dark")
tips=sns.load_dataset("tips")
sns.violinplot(x=tips["total_bill"])
sns.violinplot(x='day',y='total_bill',hue='smoker',data=tips, palette="muted")
plt.show()
```
